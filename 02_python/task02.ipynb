```python
price = int(input('税抜価格: ')) * 1.1

match price:
    case price if price >= 2000:
        print('送料は無料です')
        print(f'送料込税込価格は{int(price)}円です')
    case price if 0 <= price < 2000:
        print('送料は350円です')
        print(f'送料込税込価格は{int(price + 350)}円です')
    case _:
        print('0以上の税抜価格を入力してください')
```
