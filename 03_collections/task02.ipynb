```python
from collections import defaultdict

words = []
cnt = defaultdict(int)

while True:
    word = input('英単語を入力してください：')

    if not word:
        break

    words.append(word)
    for c in word:
        cnt[c] += 1

print('入力した英単語：' + str(sorted(words)))

for k, v in cnt.items():
    print(k + 'が' + str(v) + '個ありました')
```
