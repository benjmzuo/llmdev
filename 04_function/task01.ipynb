```python
def validate(name) -> bool:
    invalid_len = len(name) < 5 or len(name) > 15
    invalid_alnum = not name.isalnum()
    has_space = ' ' in name

    if not (invalid_len or invalid_alnum or has_space):
        return True

    if invalid_len:
        print('エラー: ユーザー名は5文字以上15文字以内で入力してください')

    if invalid_alnum:
        print('エラー: ユーザー名は英数字のみ使用できます')

    if has_space:
        print('エラー: ユーザー名に空白を含めることはできません')

name = input()
res = validate(name)

if res:
    print('ユーザー名は有効です')
else:
    print('不正なユーザー名です')
```
